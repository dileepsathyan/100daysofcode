# Auto Correlation & Auto Regressive Models

### Correlation is a measure of extend to which 2 variables are related. There can be positive or negative correlation between 2 variables which are measured by 'Correlation Coefficient' r value. It can range from -1 to 1. 

### -1 means the variables are negatively correlation and 1 shows strong positive correlation. r value of 0 indicates that there are no correlation between the variables.

### However, as the name states, 'Auto Correlation' means the correlation of a variable with its own previous values in a time series. Auto correlation can be used to predict the today's value using the values of yesterday or the previous time period.

### y(t) being the current value, y(t-1) with a lag of 1 is the previous value, y(t-2) with lag of 2 is the previous to the previous value and thus y(t-k) means the value of y with the lag of k.

## AR Models and their difference from Moving Average (MA) Models..


### AR Models assume that there exists a linear relationship between the previous values to the current value in a timeseries. However, the MA model considers that the current value of a variable is dependant on the error terms (both current and previous error terms). Since the errors are random, we can not plot a linear regression between the error terms and current value.



## What are ACF and PACF charts?


### ACF: A function that gives values of Auto-Correlation of any series with its lagged values
### PACF: An indirect function to find Auto-Correlation, after removing the relationship explained by its previous values

## Identifying the appropriate model to use for a timeseries..?


### We can use ACF and PACF charts from statsmodels to decide which model to use.

#### As in the above ACF chart, if the significant spikes are turning down to zero from either positive or negative peaks, & at the same time, if there are only 1 or 2 significant spikes in PACF charts, it is ideal to use AR model. The order at which the AR model to be used depends on the number of significant spikes in the PACF charts. 

### Also, if the significant spikes in the ACF chart are alternating positively and negatively while there keeping only few spikes in the PCAF charts, we should still consider the AR model.


### Alternatively, if the spikes are in opposite to what's mentioned above, (more spikes in PACF and lesser in ACF), we should consider MA model.


### There can be situations where both above situations can be present like in the below charts. In such situaitons, it is advisable to use ARMA models


```python
# Using the ACF and PACF charts in action

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf, pacf
```

```python
series = pd.read_csv('/Users/dileepsathyan/Documents/GitHub/datasets/ts_sales_data.csv', 
                     parse_dates=['date'],
                    index_col=0)
series.head()
```

```python
# Plot the ACF and PCAF charts

acf_lag = acf(series, nlags=100)
plt.plot(acf_lag)
```
